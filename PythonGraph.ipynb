```python
import pandas as pd
import matplotlib.pyplot as plt

# קריאת הקובץ
df = pd.read_csv('reservationsN.csv')

# פונקציה שמחזירה את סוג השהות לפי הזמנה
def get_stay_type(row):
    if row['weeknights'] > 0 and row['weekendnights'] == 0:
        return 'Only Weeknights'
    elif row['weekendnights'] > 0 and row['weeknights'] == 0:
        return 'Only Weekendnights'
    elif row['weeknights'] > 0 and row['weekendnights'] > 0:
        return 'Both'


df['StayType'] = df.apply(get_stay_type, axis=1)

counts = df['StayType'].value_counts()

plt.bar(counts.index, counts.values, color='skyblue')
plt.title('Reservations by Stay Type')
plt.xlabel('Stay Type')
plt.ylabel('Number of Reservations')

#  סכום הזמנות של כל עמודה
for i, val in enumerate(counts.values):
    plt.text(i, val + 1, str(val), ha='center')

plt.tight_layout()
plt.show()
```
